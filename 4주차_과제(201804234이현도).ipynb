### 1. movie 데이터의 크기를 줄여봅시다.
___

```python
movie.info()
```

```python
movie1 = movie.iloc[:,:11].fillna(0)
```

```python
movie1.info()
```

```python
movie1.memory_usage(deep=True)
```

```python
#우선 수치형 데이터 부터 value에 적당한 크기의 type로 변형시켜줬다.
movie1['num_critic_for_reviews']=movie1['num_critic_for_reviews'].astype(np.float16)
movie1['duration']=movie1['duration'].astype(np.float16)
movie1['director_facebook_likes']=movie1['director_facebook_likes'].astype(np.float16)
movie1['actor_3_facebook_likes']=movie1['actor_3_facebook_likes'].astype(np.float16)
movie1['actor_1_facebook_likes']=movie1['actor_1_facebook_likes'].astype(np.float32)
movie1['gross']=movie1['gross'].astype(np.float32)
```

```python
movie1.memory_usage(deep=True)
```

```python
#object의 객체 개수 확인 -> color와 genres 정도만 category로 변형해주면 되겠다고 판단.
movie1.select_dtypes(include=['object']).nunique()
```

```python
movie1['color']=movie1['color'].astype('category')
movie1['genres']=movie1['genres'].astype('category')
```

```python
movie1.memory_usage(deep=True)
```

```python
facebook_likes = movie1.filter(like='facebook_likes').sum(axis=1)/3
facebook_likes_list = list(facebook_likes)
```

```python
movie1 = movie1.assign(facebook_likes = facebook_likes_list)
```

### 2. college에서 UGDS의 최대값을 구하는데 도시별로.
___

```python
college.head()
#INSTNM, CITY, UGDS
```

```python
#오로지 groupby와 nlargest로만 구성
c1 = college[['INSTNM','CITY','UGDS']].groupby('CITY')["UGDS"].nlargest(1)[::-1]
c1
```

```python
#sort_values와 drop_duplicates로만 구성
(
college[['INSTNM','CITY','UGDS']]
.sort_values(['CITY','UGDS'],ascending=False)
.drop_duplicates(subset='CITY')
)
```

```python
#groupby와 apply(lambda sort_values)로 구성
college[['INSTNM','CITY','UGDS']].groupby('CITY',as_index=False).apply(lambda df: df.sort_values( 
    'UGDS',ascending=False 
    ).head(1)
).droplevel(0).sort_values('CITY',ascending=False) 
```

```python
#groupby와 apply(lambda nlargest)후 sort_values로 구성
(
college[['INSTNM','CITY','UGDS']]
.groupby('CITY',as_index=False)
.apply(lambda df: df.nlargest(1,'UGDS')
).droplevel(0).sort_values('CITY',ascending=False)
)
```

### 3. 주별 RELAFFIL별 UGDS의 최대값을 구하라  
   INSTNM, STABBR, RELAFFIL, UGDS
___

```python
#sort_values와 drop_duplicates로 구성
(
college[['INSTNM','STABBR','RELAFFIL','UGDS']]
.sort_values(['STABBR','RELAFFIL','UGDS'],ascending=False)
.drop_duplicates(subset=['STABBR','RELAFFIL'])
).sort_values('STABBR',ascending=True)
```

```python
#groupby와 apply(lambda sort_Values)후 sort_values로 구성
(
college[['INSTNM','STABBR','RELAFFIL','UGDS']]
.groupby(['STABBR','RELAFFIL'],as_index=False)
.apply(lambda df: df.sort_values('UGDS',ascending=False).head(1))
.droplevel(0).sort_values('STABBR',ascending=True)
)
```

```python
#groupby와 apply(lambda nlargest)후 sort_values로 구성
(
college[['INSTNM','STABBR','RELAFFIL','UGDS']]
.groupby(['STABBR','RELAFFIL'],as_index=False)
.apply(lambda df: df.nlargest(1,'UGDS'))
.droplevel(0).sort_values('STABBR',ascending=True)
)
```
