```python
movies2 = pd.read_csv('data/movie.csv',
                       dtype = {
                           'num_critic_for_reviews':np.float16,
                           'duration':np.float16,
                           'director_facebook_likes':np.float16,
                           'actor_3_facebook_likes':np.float16,
                           'actor_1_facebook_likes':np.float32,
                           'gross':np.float32,
                           'facenumber_in_poster':np.float16,
                           'num_user_for_reviews':np.float16,
                           'budget':np.float32,
                           'title_year':np.float32,
                           'actor_2_facebook_likes':np.float32,
                           'imdb_score':np.float16,
                           'aspect_ratio':np.float16,
                           'num_voted_users':np.int16,
                           'movie_facebook_likes': np.int32,
                           'cast_total_facebook_likes': np.int32,
                           'color': 'category',
                           'language': 'category',
                           'country': 'category',
                           'content_rating':'category'
                       })
movies2.info()
```

```python
facebook_likes = movies.filter(like='facebook_likes').sum(axis=1)/6
facebook_likes_list = list(facebook_likes)
```

```python
movies2 = movies2.assign(facebook_likes = facebook_likes_list)
```

```python
movies2.info()
```

----
column 이름을 보면 facebook_likes가 속해있는 열들이 많다.

하나의 평균값으로 표현함으로써 column의 개수를 줄여보자.

____

```python
list(movies.filter(like='facebook_likes').columns)
```

```python
movies2.drop(list(movies.filter(like='facebook_likes').columns),axis=1,inplace=True)
```

```python
movies2.info()
```

```python
movies2 = movies2.astype({'facebook_likes':'float32'})
```

```python
movies2.info()
```

```python
movies2.to_csv('movies2.csv')
```

```python
new_movies = pd.read_csv('movies2.csv',
                         dtype={
                           'num_critic_for_reviews':np.float16,
                           'duration':np.float16,
                           'gross':np.float32,
                           'facenumber_in_poster':np.float16,
                           'num_user_for_reviews':np.float16,
                           'budget':np.float32,
                           'title_year':np.float32,
                           'imdb_score':np.float16,
                           'aspect_ratio':np.float16,
                           'num_voted_users':np.int16,
                           'color': 'category',
                           'language': 'category',
                           'country': 'category',
                           'content_rating':'category',
                           'faceboook_likes': np.float32
                         },index_col=0,
                           chunksize = 301)

def process(df):
    return f'processed {df.shape} shape of df {df.iloc[0,:]}'

for chunk in new_movies:
    print(process(chunk),"\n")
```
